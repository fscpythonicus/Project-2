```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('james_bond_data_cleansed.csv')
```

```python
plt.scatter(data["imdb"], data["rotten_tomatoes"])
plt.xlabel('IMDb')
plt.ylabel('Rotten Tomatoes')
plt.title('IMDb vs Rotten Tomatoes Ratings')
plt.show()
```

There seems to be a positive correlation between the IMDb and Rotten Tomatoes ratings of the James Bond films, as the plotted data roughly creates the shape of a line.

```python
plt.scatter(data["imdb"], data["bond_kills"])
plt.xlabel('IMDb')
plt.ylabel('Kills')
plt.title('IMDb Ratings vs James Bond Kills')
plt.show()
```

There does not seem to be a correlation between the IMDb Ratings and the amount of kills James Bond makes in the films.

```python
plt.plot(data["movie_budget"],"b-o")
plt.plot(data["income_world"],"r-o")
plt.legend(["Movie Budget", "Global Income"])
plt.title('Movie Budget vs Global Income')
plt.show()
```

As the global income of the movies increased, more money was put into the budget of the movies over time. With each significant spike in global income, the movie budget soon rises to match.

```python
plt.figure(figsize=(12, 6))
data['budget_to_income_ratio'] = data['income_world'] / data['movie_budget']
plt.bar(x=data['release_year'], height=data['budget_to_income_ratio'], width=0.8)
plt.xlabel('Year Released')
plt.ylabel('Income to Budget Ratio')
plt.title('Global Income to Movie Budget Ratio Over Time')
plt.show()
```

Although the movies are making more money than in previous years, the ratio has decreased as the budget for movie production has increased. It is an overall net positive, but the increasing budget for movie production over the years has negatively impacted the income:budget ratio.

```python
film_length = data['film_length']
length_sorted = film_length.sort_values(ascending=[True])
print(length_sorted)

length_mean = film_length.mean()
print(f"The mean of film lengths is: {length_mean} minutes")

length_median = film_length.median()
print(f"The median of film lengths is: {length_median} minutes")
```

From the sorted list, you can see that the shortest film is film 21, Quantum of Solace. The longest is 24, the last film, No Time to Die.
You can also see that the mean of the film lengths is 128.28 minutes, with a median of 128 minutes.

```python
film_rating = (data['imdb'] + data['rotten_tomatoes']) / 2
rating_sorted = film_rating.sort_values(ascending=[True])
print(rating_sorted)

rating_mean = film_rating.mean()
print(f"The mean of film ratings is: {rating_mean}/10")

rating_median = film_rating.median()
print(f"The median of film ratings is: {rating_median}/10")
```

The lowest rated film is film 13, A View to Kill. The highest rated is film 2, Goldfinger.
The mean of the films' ratings, calculated by adding the IMDb and Rotten Tomatoes ratings and averaging the result, is a 6.694/10. The median of the film ratings is a 6.55/10.
